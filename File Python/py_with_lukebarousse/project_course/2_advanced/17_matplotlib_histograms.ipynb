```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# convert to datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
# convert the string representation of list to actual list
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_ID = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == "Indonesia")].copy()
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == "United States")].copy()
```

```python
df_DA_ID['salary_year_avg'].plot(kind='hist', bins=30, edgecolor='black')
# plt.xlim(0, 120000)

ax = plt.gca()

ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
plt.title("Distribution of Indonesia Data Analyst Yearly Salaries")
plt.ylabel("Number of Jobs")
plt.xlabel("Yearly Salaries")
plt.show()
```

```python
df_DA_US['salary_year_avg'].plot(kind='hist', bins=30, edgecolor='black')
# plt.xlim(0, 250000)

ax = plt.gca()

ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
plt.title("Distribution of United States Data Analyst Yearly Salaries")
plt.ylabel("Number of Jobs")
plt.xlabel("Yearly Salaries")
plt.show()
```
