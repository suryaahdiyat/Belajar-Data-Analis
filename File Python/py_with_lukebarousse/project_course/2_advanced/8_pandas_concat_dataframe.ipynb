```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# convert to datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b') # extract month name
```

```python
months = df['job_posted_month'].unique() # get unique months
# create dictionary of dataframes, one for each month, key is month name
dict_months = {month: df[df['job_posted_month'] == month] for month in months}
# dict_months['Jan']
```

```python
df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)
```

```python
df_q1['job_posted_month'].value_counts().plot(kind='bar')
plt.xticks(rotation=0)
```
