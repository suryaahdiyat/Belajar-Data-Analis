```python
class JobPost:
  def __init__(self, title, location, salary):
    self.title = title
    self.location = location
    self.salary = salary

  def display_info(self):
    print(f"Job Title:    {self.title}")
    print(f"Location:     {self.location}")
    print(f"Salary:       {self.salary:,.0f}")
```

```python
job_1 = JobPost(job_title, job_location, job_salary)
```

```python
job_1.display_info()
```

```python
years_experients = {
  'Luke' : 2,
  'Kelly' : 4,
  'Surya' : 5,
  'Alex' : 3
}
```

```python
for person in years_experients:
  print(person)
```

```python
for name, year in years_experients.items():
  if year > 3:
    print(name, year)
```

```python
years = 0
index = 0

years_list = list(years_experients.items())
while years < 3:
  key, years = years_list[index]
  if years > 3:
    print(key, years)
  index += 1
```
