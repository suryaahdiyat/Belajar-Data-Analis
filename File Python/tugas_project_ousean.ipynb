```python
import pandas as pd
```

```python
def fixColumnSpace(lists): 
  newList = []
  for i, c in enumerate(lists):
    if " " in c:
      list_of_items = list(c)
      if list_of_items[len(list_of_items) -1] == " ":
        list_of_items.pop()
      newList.append(''.join(list_of_items))
    else:
      newList.append(c)
      
  return newList
```

```python
def replaceSpace(lists):
  newList = []
  for i, c in enumerate(lists):
    if " " in c:
      newList.append(c.replace(" ", "_"))
    else :
      newList.append(c)
  return newList
```

```python
df = pd.read_excel('../File excel/bootcamp_ousean/Form Pendaftaran Internship (Jawaban).xlsx')

broken_column = df.columns

column_fixed = fixColumnSpace(broken_column)

column_fixed = replaceSpace(column_fixed)

df.columns = column_fixed
# print(replaceSpace(broken_column))
# print("++++++++")
# print(replaceSpace(column_fixed))
```
